```python
import cv2
from app.steps import ScreenshotProcessor
from app.adb import ADB

working_dir = "app/screenshots/"
img_name = input("Image name: ")

adb = ADB("emulator-5558")
adb.ScreenCapture(working_dir, img_name)
# Переменные для хранения начальных и конечных координат
drawing = False  # True, когда мышь нажата
start_point = (-1, -1)  # Начальная точка
end_point = (-1, -1)  # Конечная точка

# Функция для обработки событий мыши
def draw_rectangle(event, x, y, flags, param):
    global start_point, end_point, drawing

    if event == cv2.EVENT_LBUTTONDOWN:  # ЛКМ нажата
        drawing = True
        start_point = (x, y)

    elif event == cv2.EVENT_MOUSEMOVE:  # Мышь перемещается
        if drawing:
            end_point = (x, y)

    elif event == cv2.EVENT_LBUTTONUP:  # ЛКМ отпущена
        drawing = False
        end_point = (x, y)

# Открываем изображение с помощью OpenCV

screenshot = cv2.imread(f"{working_dir}{img_name}.png")

# Копия изображения для отображения
image_copy = screenshot.copy()

# Назначаем функцию обратного вызова для окна
cv2.namedWindow('Draw Rectangle')
cv2.setMouseCallback('Draw Rectangle', draw_rectangle)

while True:
    # Копируем изображение для отрисовки прямоугольника
    img_to_show = image_copy.copy()

    # Если есть конечная точка, рисуем прямоугольник
    if start_point != (-1, -1) and end_point != (-1, -1):
        cv2.rectangle(img_to_show, start_point, end_point, (0, 255, 0), 2)

    # Показываем изображение
    cv2.imshow('Draw Rectangle', img_to_show)

    # Если нажать 'q', программа завершится
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        break

# Вычисляем ширину и высоту прямоугольника
x, y = start_point
w = abs(end_point[0] - start_point[0])
h = abs(end_point[1] - start_point[1])

coordinates = [x,y,w,h]

ScreenshotProcessor().crop(f"{working_dir}{img_name}.png", coordinates, f"{working_dir}{img_name}.png")
# Закрываем все окна
print(coordinates)
cv2.destroyAllWindows()
```

```python
import yaml
import logging
from dataclasses import dataclass
from dataclasses import dataclass, field
from typing import Optional, List
from abc import ABC, abstractmethod
from enum import Enum

from app.adb import ADB
from app.adb import LDConnector
from config import load_config


class OnFalseEnums(Enum):
    skip = "skip"
    retry = "retry"

class StepTypesEnums(Enum):
    button = "button"
    input_filed = "input_field"
    validator = "validator"

class WorkSpaceArea():
    def __call__(self, workspace_area):
        if sum(workspace_area) > 0:
            x, y, w, h = workspace_area
            return f"{x}x{y}+{w}+{h}"
        return []


class Step(ABC):
    def __init__(self, step_name: str, screenshot_name: str, required_screen_check: bool = False, threshold: float = 0.75, on_false: OnFalseEnums = OnFalseEnums.retry, retries_limit: int = 2, workspace_area = []):
        self.step_name = step_name
        self.screenshot_name = screenshot_name
        self.required_screen_check = required_screen_check
        self.threshold = threshold
        self.retries_limit = retries_limit
        self.workspace_area = workspace_area
        self.on_false = on_false
        if on_false:
            self.on_false = OnFalseEnums(on_false)
    
    def validate_workspace_area(self):
        if WorkSpaceArea()(self.workspace_area):
            return True
        return False
    
    @abstractmethod
    def default_execute(self):
        pass
    
    
class ClickStep(Step):
    def __init__(self, step_name: str, screenshot_name:str, type, workspace_area = [], required_screen_check = False, threshold = 0.75, on_false = OnFalseEnums.retry, retries_limit = 2):
        super().__init__(step_name, screenshot_name, required_screen_check, threshold, on_false, retries_limit)
        self.workspace_area = workspace_area
        self._func = None
        self.type = StepTypesEnums(type)



    def click_handler(self, func):
        def wrapper(adb, step, *args, **kwargs):
            # Вызываем декорированную функцию с adb, step, args и kwargs
            return func(adb, step, *args, **kwargs)
        
        self._func = wrapper  # Сохраняем обертку функции
        return wrapper  # Возвращаем обертку функции  

    def __call__(self, *args, **kwargs):
        if self._func:
            return self._func(*args, **kwargs)
        else:
            return self.default_execute(*args, **kwargs)
        
    def default_execute(self, adb: ADB):
        adb.random_click(self.workspace_area)
        

class InputStep(Step):
    def __init__(self, step_name, screenshot_name, type, select_placeholder: bool = False, required_screen_check = False, threshold = 0.75, on_false = OnFalseEnums.retry, retries_limit = 2, workspace_area=[]):
        super().__init__(step_name, screenshot_name, required_screen_check, threshold, on_false, retries_limit, workspace_area)
        self.select_placeholder = select_placeholder
        if self.select_placeholder and not self.validate_workspace_area:
            return ValueError(f"worspace_area field needed for {self.step_name}")
        self.workspace_area = workspace_area
        self._func = None
        self.type = StepTypesEnums(type)

    def __call__(self, *args, **kwargs):
        if self._func:
            return self._func(*args, **kwargs)
        else:
            return self.default_execute(*args, **kwargs)

    def default_execute(self, adb: ADB):
        if self.select_placeholder:
            # if self.validate_workspace_area():
            adb.random_click(self.workspace_area)

        adb.input_text(adb.emulator_name, self.step_name)

class ValidatorStep(Step):
    def __init__(self, step_name, screenshot_name, type, required_screen_check = False, threshold = 0.75, on_false = OnFalseEnums.retry, retries_limit = 2, workspace_area=[]):
        super().__init__(step_name, screenshot_name, required_screen_check, threshold, on_false, retries_limit, workspace_area)

        self.type = StepTypesEnums(type)

    

    def __call__(self, *args, **kwargs):
        if self._func:
            return self._func(*args, **kwargs)
        else:
            raise ValueError("Function is not declired")

    def default_execute(self, adb: ADB):
       pass

class GetterStep(Step):
    def __init__(self, step_name, screenshot_name, required_screen_check = False, threshold = 0.75, on_false = OnFalseEnums.retry, retries_limit = 2, workspace_area=[]):
        super().__init__(step_name, screenshot_name, required_screen_check, threshold, on_false, retries_limit, workspace_area)


# @dataclass
# class StepData:
#     step_name: str
#     screenshot_name: str
#     type: str 
#     required_screen_check: bool = False
#     threshold: Optional[float] = 0.75
#     on_false: Optional[str] = None  
#     select_placeholder: Optional[bool] = None 
#     workspace_area: List[int] = field(default_factory=list)

@dataclass(init=False)
class StepData:
    step_name: str
    screenshot_name: str
    type: str
    extra_args: list = field(default_factory=list)
    extra_kwargs: dict = field(default_factory=dict)

    def __init__(self, step_name: str, screenshot_name: str, type: str, *args, **kwargs):
        self.step_name = step_name
        self.screenshot_name = screenshot_name
        self.type = type
        self.extra_args = args
        self.extra_kwargs = kwargs
        

    # def __post_init__(self, *args, **kwargs):
    #     self.extra_args = args
    #     self.extra_kwargs = kwargs

class StepManager():
    def __init__(self, config: str):
        self.adb = ADB(emulator_name="emulator-5554")
        self.ldplayer = LDConnector()
        self.emulator_name = "emulator-5554"
        self.steps = []
        self.config = load_config(config)
        self.screenshots_path = self.config.get("screenshots_path")
        self.work_dir = self.config.get("work_dir")

        for step in self.config.get("steps"):
            data = StepData(**step)

            if data.type == StepTypesEnums.button.value:
                self.steps.append(ClickStep(**step))

            if data.type == StepTypesEnums.input_filed.value:
                self.steps.append(InputStep(**step))

            if data.type == StepTypesEnums.validator.value:
                self.steps.append(ValidatorStep(**step))

    def validate_screenshot(self, step: Step, attemp_delay = 0.25):
        attemps = 1
        print(f"Validating {step.step_name}")
        print(step.on_false)
        while attemps <= step.retries_limit:
            self.adb.ScreenCapture(self.adb.emulator_name, self.work_dir, self.emulator_name)
            result = find_image_in_screenshot(f"{self.screenshots_path}{step.screenshot_name}", f"{self.work_dir}{self.emulator_name}.png", step.threshold)

            if not result and step.on_false == OnFalseEnums.skip:
                logging.debug(f"Skipping {step.step_name}")
                return OnFalseEnums.skip

            if not result:
                logging.debug(f"State of {step.step_name} screenshot not valid, trying again. Retries {attemps}")
                attemps += 1
                logging.debug(attemp_delay)
                continue

            if result:
                logging.debug(f"Valide state of {step.step_name} after {attemps} retries")
                return True

        return False

    def set_step_handler(self, step_name):
        for step in self.steps:
            if step.step_name == step_name:
                def wrapper(func):
                    def inner_wrapper(adb, *args, **kwargs):
                        # Вызываем декорированную функцию с adb, args и kwargs
                        return func(adb, step, *args, **kwargs)

                    step._func = inner_wrapper  # Сохраняем обертку функции в шаге
                    return inner_wrapper  # Возвращаем обертку функции

                return wrapper  # Возвращаем обертку для функции

        raise ValueError(f"Step with name '{step_name}' not found.")

        


    def execute_step(self, step: Step, *args, **kwargs):
        print(step.step_name)
        if hasattr(step, "__call__"):
            return step(self.adb, *args, **kwargs)
        
        if hasattr(step, "default_execute"):
            return step.default_execute(self.adb)
        
        raise ValueError(f"{step.step_name} has no execute options")
        

    def start(self, delay = 0.3, start_from: str = None):
        active_steps = self.steps
        if start_from:
            for k in range(0, len(active_steps)-1):
                if active_steps[k].step_name == start_from:
                    active_steps = active_steps[k::]
                    break

        for step in active_steps:
            time.sleep(delay)
            if step.required_screen_check:
                is_valid = self.validate_screenshot(step)

                if is_valid: 
                    self.execute_step(step = step)

                if is_valid == OnFalseEnums.skip:
                    print(f"Skipping {step.step_name}")
                    continue

                if not is_valid:
                    return f"Error on {step.step_name}, {is_valid}"
            
            else:
                self.execute_step(step = step)    
            
           

import time

sm = StepManager("step_maps/racer-pointx.yml")

@sm.set_step_handler("input_phone_number")
def phone_number(adb: ADB, *args, **kwargs):
    adb.input_text(adb.emulator_name, 48791727110)

@sm.set_step_handler("find_bot")
def find_bot(adb: ADB, step: Step, *args, **kwargs):
    adb.random_click(step.workspace_area)
    adb.input_text(adb.emulator_name, "@OKX_official_bot")

@sm.set_step_handler("input_confirmation_code")
def input_code(adb: ADB, *args, **kwargs):
    adb.input_text(adb.emulator_name, input("Code: "))

@sm.set_step_handler("click_on_bot_chat")
def click_on(adb: ADB, step: Step, *args, **kwargs):
    adb.random_click(step.workspace_area)

sm.start(start_from="commit_client_terms")
```

```python
import numpy
import subprocess
def ScreenCaptureNoSave(workspace_area: list[int]):
    # Захватываем скриншот и возвращаем его как байты
    pipe = subprocess.Popen("adb.exe -s emulator-5554 exec-out screencap -p", stdout=subprocess.PIPE, shell=True)
    img_bytes = pipe.stdout.read()

    # Преобразуем байты в numpy массив, который можно использовать в OpenCV
    img_array = numpy.frombuffer(img_bytes, dtype=numpy.uint8)

    # Декодируем изображение из массива байт (массив нужно преобразовать в PNG или другое изображение)
    screenshot = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
    if workspace_area:
        x,y,w,h = workspace_area
        return screenshot[y:y + h, x:x + w]
    return screenshot

def load_template_as_bytes(template_path):
    # Загрузим шаблонное изображение с диска как numpy массив (не сохраняем на диск)
    template = cv2.imread(template_path, cv2.IMREAD_COLOR)
    return template

def find_image_in_screen(screenshot, template, threshold=0.8):
    # Сравниваем шаблон с скриншотом используя метод matchTemplate
    result = cv2.matchTemplate(screenshot, template, cv2.TM_CCOEFF_NORMED)
    
    # Определяем, где совпадения выше порога (threshold)
    loc = numpy.where(result >= threshold)

    matched_coordinates = []
    for pt in zip(*loc[::-1]):  # Инвертируем координаты
        matched_coordinates.append((pt[0], pt[1]))

    if matched_coordinates:
        return True  # Если совпадение найдено
    return False

# screen = ScreenCaptureNoSave([303,270,300,210])
# ADB("emulator-5554").ScreenCapture("emulator-5554", "screenshots/trash/", "test_text_reading")
temp = load_template_as_bytes("screenshots/trash/test_text_reading.png")
x,y,w,h = 303,270,300,210
temp = temp[y:y + h, x:x + w]

```

```python
import cv2
from cv2.mat_wrapper import Mat as MatLike
from enum import Enum
import numpy
# from ABC import abstractmethod



class ColorTypes(Enum):
    BGR = "BGR"
    BINARY = "BINARY"
    GRAYSCALE = "GRAYSCALE"
    UNKNOWN = "UNKNOWN"

class BitwiseScreenshotProcessor(ABC):

    @abstractmethod
    def test(self, path: str = "//"):
        pass

    def check_image_format(self, image):
        # if not isinstance(image, numpy.ndarray):
        #     image = numpy.ndarray(image[0][0])

        if len(image.shape) == 3 and image.shape[2] == 3:
            return ColorTypes.BGR
        
        elif len(image.shape) == 2:
            unique_values = numpy.unique(image)
            if set(unique_values).issubset({0, 255}):
                return ColorTypes.BINARY
            else:
                return ColorTypes.GRAYSCALE
        else:
            return ColorTypes.UNKNOWN
    
    def screenshot_to_bitwise(self, template_path: str, convert_to_grayscale: bool = False, convert_to_binary: bool = False):
        bitwise_screenshot = cv2.imread(template_path, cv2.IMREAD_COLOR)

        if convert_to_grayscale:
            bitwise_screenshot = cv2.cvtColor(bitwise_screenshot, cv2.COLOR_BGR2GRAY)

        if convert_to_binary:
            # Преобразование изображения в черно-белое (бинарное)
            _, bitwise_screenshot = cv2.threshold(bitwise_screenshot, 127, 255, cv2.THRESH_BINARY)
        
        return bitwise_screenshot
    
    def bitwise_find_image_in_screen(self, screenshot: MatLike, template: MatLike | str, threshold=0.8):

        if isinstance(template, str):
            screenshot_type = self.check_image_format(screenshot)
            if screenshot_type == ColorTypes.BGR:
                template = self.screenshot_to_bitwise(template)
            
            if screenshot_type == ColorTypes.BINARY:
                template = self.screenshot_to_bitwise(template, convert_to_binary=True, convert_to_grayscale=True)

            if screenshot_type == ColorTypes.GRAYSCALE:
                template = self.screenshot_to_bitwise(template, convert_to_grayscale=True)
            

        result = cv2.matchTemplate(screenshot, template, cv2.TM_CCOEFF_NORMED)
        
        location = numpy.where(result >= threshold)

        matched_coordinates = []
        for pt in zip(*location[::-1]):
            matched_coordinates.append((pt[0], pt[1]))

        if matched_coordinates:
            return True 
        return False
    
    def bitwise_crop(self, screenshot: MatLike, coordinates: list[int]):
        x, y, w, h = coordinates
        return screenshot[y:y + h, x:x + w]
    

class tst(BitwiseScreenshotProcessor):
    x = 123
    def test(self, path):
        print(path)

hasattr(tst, "x")
```
